```python
MonthsAndDays = {"January":31, "February":28, "March":31,\
                 "April":30, "May":31, "June":30,\
                 "July":31, "August":31, "September":30,\
                 "October":31, "November":30, "December":31}
```

```python
def week(day):
    
    Days_in_week = ["Sunday", "Monday", "Tuesday", 'Wednesday', 'Thursday', 'Friday', 'Saturday']
    
    Day_No=ind=Days_in_week.index(day)
    
    while ind != len(Days_in_week)+Day_No:
        yield Days_in_week[ind%7]
        ind += 1
```

```python
def last_day_of_a_month(week, date): return week[date%7]
```

```python
last_day_of_a_month(list(week('Monday')), 31)
```
